```python
import pandas as pd
```

```python
train = pd.read_csv("./train.csv")
```

```python
train
```

```python
train["outcome"].unique()
```

```python
outcome = {"died":0, "euthanized":1, "lived":2}
```

```python
train["outcome"] = train["outcome"].map(outcome)
```

```python
train
```

```python
s = (train.dtypes == 'object')
object_cols = list(s[s].index)
print(object_cols)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ordiner_encoder = OrdinalEncoder()
```

```python
label_train = train.copy()
label_train[object_cols] = ordiner_encoder.fit_transform(train[object_cols])
```

```python
label_train
```

```python
label_train = label_train.drop("id", axis=1)
```

```python
label_train
```

```python
idx = label_train.columns
corr = label_train.corr()
```

```python
corr["outcome"].abs()
```

```python
from matplotlib import pyplot as plt
```

```python
plt.barh(corr["outcome"].abs().index, corr["outcome"].abs().values)
plt.grid()
plt.show()
```

```python
for i in idx:
    s = (corr[i].abs() >= 0.2).sum()
    if s > 5:
        print(i)
```
